# Supervised Learning

The supervised learning is a machine learning technique which the model is trained with a labeled dataset, meaning the data includes both inputs and the desired outputs

- The predicted values are knwown
- Supervised learning uses labels
- Aim: Predict the target values of unseen data, given the features

![Supervised learning example](https://static.javatpoint.com/tutorial/machine-learning/images/supervised-machine-learning.png "Supervised learning example")

## Types of supervised learning

The supervised learning is divided mainly into two types of supervised learning, classification and regression

### Classification

They are used to predict discrete / categorical outcomes (y) using independent variables (x)

More info: [Classification models](./classification/classification.ipynb)

#### Classification models

Some of most common models are:

- [k-Nearest Neighbors](./classification/k-Nearest-Neighbors.ipynb)
- [Logistic Regression](./classification/logistic-regression.ipynb)
- [Decision Trees](./classification/decision-trees.ipynb)
- [Random Forest](./classification/random-forest.ipynb)
- [Support Vector Machines (SVM)](./classification/svm.ipynb)
- [Neural Networks](./classification/neural-networks.ipynb)

#### Steps

1. Build a classifier model
2. Model learns from the labeled data we pass to it
3. Pass unlabeled data to the model as input
4. Model predicts the labels of the unseen data

### Regression

The regression is used to predict a continuous target variable

#### Regression Models

Some of most common models are:

- [Linear Regression](./regression/linear-regression.ipynb)
- [Polynomial Regression](./regression/polynomial-regression.ipynb)
- [Ridge and Lasso Regression](./regression/ridge-lasso-regression.ipynb)
- [Decision Trees for Regression](./regression/decision-trees-regression.ipynb)
- [Random Forest for Regression](./regression/random-forest-regression.ipynb)


## Requirements

To effectively apply supervised learning techniques, certain requirements must be met in the data:

- No missing values
- Data in numeric format
- Data stored in pandas Dataframe or Numpy array
